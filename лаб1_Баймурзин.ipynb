```python
import numpy as np
import math
import random

A = np.array([[random.randint(0,100) for i in range(5)] for i in range(8)])

B = np.array([[random.randint(0,100) for i in range(8)] for i in range(5)])

# np.dot() - матричное умножение
# Важно! Количество столбцов первой матрицы должно равняться количеству строк второй
AB = np.dot(A,B)
BA = np.dot(B,A)

# np.eye() - создает единичную матрицу (единицы по диагонали, остальное нули)
# shape[0] - берет размер из первого измерения матрицы
C = 2*AB - np.eye((2*AB).shape[0])

# np.linalg.det() - вычисляет определитель матрицы
# Работает только с квадратными матрицами!
print(
    f'A*B \n {AB}\n\n',
    f'B*A \n {BA}\n\n',
    f'C \n {C}\n\n',
    f'\nОпределитель матрицы C:", {np.linalg.det(C)}\n\n'
    )
# np.linalg.inv() - находит обратную матрицу
# Может вызвать ошибку если определитель = 0
try:
    C_inv = np.linalg.inv(C)
    print("\nОбратная матрица C:")
    print(C_inv,'\n\n')
except np.linalg.LinAlgError:
    print("\nОбратная матрица не существует")

# Ищем минимумы разными способами
# Сначала через numpy - это намного проще

# np.min() с параметром axis:
# axis=1 - минимум по строкам (горизонтально)
# axis=0 - минимум по столбцам (вертикально)
# без axis - минимум по всей матрице
print(
    f'Минимумы по строкам (np):, {np.min(A, axis=1)}\n\n',
    f'Минимумы по столбцам (np):, {np.min(A, axis=0)}\n\n',
    f'Общий минимум (np):, {np.min(A)}\n\n'
    )

# Теперь ищем минимумы через циклы (старый способ)
# A[i,:] - выбирает i-ю строку целиком
# A[:,i] - выбирает i-й столбец целиком
arrayMinRows = []
for i in range(A.shape[0]):
  arrayMinRows.append(min(A[i,:]))
arrayMinCell = []
for i in range(A.shape[1]):
  arrayMinCell.append(min(A[:,i]))

minTotal = min([min(arrayMinCell), min(arrayMinRows)])

print(
    f'\nМинимумы по строкам (цикл):, {arrayMinRows}\n\n',
    f'Минимумы по столбцам (цикл):,  {arrayMinCell}\n\n',
    f'Общий минимум (цикл):, {minTotal}\n\n')

# np.sin() - применяет синус к каждому элементу матрицы("векторизация"
g_A = np.sin(A)
g_A = np.sin(A)
print("\nРезультат применения g(z) = sin(z) к матрице A:\n",g_A)

#Задача 5

# A[-3:] - срез последних трех строк матрицы
print(f'\n Срез 3 последних строк: \n{A[-3:]}\n\n',
      f'Столбцы с нечетными номерами: \n {[A[:,i] for i in range(1, A.shape[1], 2)]}\n\n'
      f'Числа соответветсвующие обоим условиям: \n {[[ el for el in A[-3:,i] ] for i in range(1, A.shape[1], 2)] }'
    )


#Задача 6
# Сначала добавляем строку с максимумами, затем добавляем столбец единиц
result = np.vstack((A, np.max(A, axis=0)))  # добавляем строку максимумов
ones_column = np.ones((result.shape[0], 1))  # ones - создает массив из единиц
final_result = np.hstack((result, ones_column)).astype(int)  # добавляем столбец справа

print(
    f'\n Матрица с максимумами по столбцам и единичным столбцом справа:\n{final_result}'
)
```

```python
import numpy as np
from scipy import optimize

# Определяем функцию
def f(x):
    return (x + 1) / ((x - 1)**2)

# Определяем производную функции для проверки результатов
def df(x):
    return ((x - 1)**2 - 2*(x + 1)*(x - 1)) / ((x - 1)**4)

# Пробуем разные начальные точки
starting_points = [-4, -2, -1, -0.5]

print("Результаты из разных начальных точек:")
print("-" * 60)

for x0 in starting_points:
    result = optimize.minimize(f, x0, method='BFGS')
    print(f"\nНачальная точка x₀ = {x0}")
    print(f"Найденный минимум в точке x = {result.x[0]:.6f}")
    print(f"Значение функции в минимуме: {result.fun:.6f}")
    print(f"Количество итераций: {result.nit}")
    print(f"Успех оптимизации: {result.success}")
    print(f"Градиент в точке минимума: {df(result.x[0]):.6e}")
```

```python
import matplotlib.pyplot as plt
import numpy as np

def f(x):
    return (x + 1) / ((x - 1)**2)
# f'
def df(x):
    return ((x - 1)**2 - 2*(x + 1)*(x - 1)) / ((x - 1)**4)

# plt.subplots() создает фигуру и массив подграфиков
# figsize задает размер всей фигуры
# 2, 3 означает 2 строки и 3 столбца подграфиков
fig, ((ax1, ax2, ax3), (ax4, ax5, ax6)) = plt.subplots(2, 3, figsize=(15, 10))

# Создаю массивы значений x с помощью np.linspace
# np.linspace создает равномерно распределенные точки между start и end
x1 = np.linspace(-5, 0, 1000)     # Для левой части графика
x2 = np.linspace(0.7, 0.95, 1000) # Для области около особой точки
x3 = np.linspace(1.05, 5, 1000)   # Для правой части графика

# Задаю стили линий через словари - так удобнее переиспользовать
# color - цвет линии
# linewidth - толщина линии
# linestyle - тип линии (сплошная/пунктирная и т.д.)
function_style = {
    'color': 'blue',
    'linewidth': 2,
    'linestyle': '-'
}

derivative_style = {
    'color': 'red',
    'linewidth': 1.5,
    'linestyle': '--'
}

# plot() рисует линию
# set_title() задает заголовок
# set_xlabel/ylabel() подписывают оси
# grid() включает сетку
ax1.plot(x1, f(x1), **function_style)  # ** распаковывает словарь со стилями
ax1.set_title('Левая часть графика')
ax1.set_xlabel('x')
ax1.set_ylabel('f(x)')
ax1.grid(True)

ax2.plot(x2, f(x2), **function_style)
ax2.set_title('Область около x=1')
ax2.set_xlabel('x')
ax2.set_ylabel('f(x)')
ax2.grid(True)

ax3.plot(x3, f(x3), **function_style)
ax3.set_title('Правая часть графика')
ax3.set_xlabel('x')
ax3.set_ylabel('f(x)')
ax3.grid(True)

# Строю графики производной
ax4.plot(x1, df(x1), **derivative_style)
ax4.set_title('Производная (левая часть)')
ax4.set_xlabel('x')
ax4.set_ylabel('f\'(x)')
ax4.grid(True)

ax5.plot(x2, df(x2), **derivative_style)
ax5.set_title('Производная около x=1')
ax5.set_xlabel('x')
ax5.set_ylabel('f\'(x)')
ax5.grid(True)

ax6.plot(x3, df(x3), **derivative_style)
ax6.set_title('Производная (правая часть)')
ax6.set_xlabel('x')
ax6.set_ylabel('f\'(x)')
ax6.grid(True)
# tight_layout() автоматически регулирует расположение графиков
# чтобы они не накладывались друг на друга
plt.tight_layout()
plt.show()  # Отображает все графики

"""
1. Что я узнал о библиотеке Matplotlib:
   - plt.subplots() - функция для создания нескольких графиков
   - plot() - основной метод для построения линий
   - Можно настраивать практически всё: цвета, стили, толщину, подписи
   - Графики можно организовать в виде таблицы (матрицы)

2. Основные методы, которые я использовал:
   - plot() - рисует линию
   - set_title() - задает название графика
   - set_xlabel(), set_ylabel() - подписи осей
   - grid() - добавляет сетку
   - tight_layout() - автоматическое размещение
   - show() - показывает результат

3. Почему я выбрал такие настройки:
   - Синий цвет для функции - классический выбор, хорошо видно
   - Красный для производной - контрастирует с синим
   - Пунктир для производной - чтобы отличать от основной функции
   - Толщина линий: 2 для функции и 1.5 для производной - так лучше видно

4. Почему такие диапазоны:
   - [-5, 0]: тут находится минимум функции (≈ -3)
   - [0.7, 0.95]: показывает, что происходит около точки x=1 слева
   - [1.05, 5]: показывает поведение после точки

4. Что показывают графики:
   - Верхний ряд (функция):
     * Слева видно минимум в точке x ≈ -3
     * В середине видно, как функция стремится к +∞ при x → 1-
     * Справа видно, как функция убывает после x = 1

   - Нижний ряд (производная):
     * Слева производная пересекает ось x в точке минимума
     * В середине и справа видно резкое изменение около x = 1

"""
```
